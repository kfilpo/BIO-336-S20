## Deer difusion

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
for t in range(n_animals):
        if x[t] >= 1:    x[t] = 1
        elif x[t] <= 0:  x[t] = 0
        elif y[t] >= 1:  y[t] = 1
        elif y[t] <= 0:  y[t] = 0
        else:            pass
```

```python
%matplotlib notebook

n_animals = 100
n_steps = 1000
diffusion_speed = 0.01

x = np.ones(n_animals)*0.5
y = np.ones(n_animals)*0.5

fig = plt.figure(figsize = (3,3));
plt.xlim(0,1);
plt.ylim(0,1);
#plt.show(); not needed 

for i in range(n_steps):
    x += np.random.randn(n_animals)*diffusion_speed
    y += np.random.randn(n_animals)*diffusion_speed
    np.maximum(x,0)
    np.minimum(x,1)
    np.maximum(y,0)
    np.minimum(y,1)
                
    plt.cla()
    plt.plot(x, y, 'k.', markersize = 1);
    plt.xlim(0,1);
    plt.ylim(0,1);
    fig.canvas.draw();
    time.sleep(0.001)
    
```
