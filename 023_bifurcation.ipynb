# Bifurcation model

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
nsteps = 100 
fig = plt.figure(figsize=(9,9));
x = 0.67 
i = 0
for r in np.arange(2.8, 4, 0.001):
    i = i+1
    for j in range(nsteps +(i%50)):
        x= r*x*(1-x)
    plt.plot(r, x, 'k.', markersize=1)
```
