# Simple numberical integration for motion

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#attempt 1 throwing stone upwards 

a = -1               #acceleration, negative bc gravity 
v = 5                #we throw stone upwards so positive
y = 0                #where we start throw
dt = 0.01            #size of step increase, numerical integration, supposed to be small
nsteps = 1000
history = np.zeros(nsteps)   # we will plot eventually
for i in range(nsteps):
    v = v+a*dt
    y = y+v*dt
    history[i]=y
    
plt.plot(history);
```

```python
#attempt 2 throwing stone angled and up

#notes: dt is parameter of model, not realy like important for the real world math

a = -1               #acceleration, negative bc gravity                 
vx = 1               #we throw stone upwards so positive
vy = 5
x = 0                #horizontal dist
y = 0                #vertical dist
dt = 0.01             #size of step increase, numerical integration, supposed to be small
nsteps = round(10/dt)
  
for i in range(nsteps):
    vy = vy+ a*dt
    x = x+ vx*dt
    y= max(y + vy*dt, 0)
    plt.plot(x,y,'k.');
```
