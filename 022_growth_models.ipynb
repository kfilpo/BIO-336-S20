# Growth of bacteria model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#exponential growth model

nsteps = 100          #number of steps in simulation
x = np.zeros(nsteps)   #number of animals
x[0] = 1
r = 1.1                #rate of population change - growth and deaths
for i in range(1,nsteps):
    x[i]= x[i-1]*r
    
plt.plot(x);
```

```python
#logistic grown model

nsteps = 100          
x = np.zeros(nsteps)   
x[0] = 0.6             #population size
r = 3.9                   #rate of population change - growth and deaths
for i in range(1,nsteps):
    x[i]= r*x[i-1]*(1-x[i-1])
    
plt.plot(x);
plt.ylim((0,1));       #make axis between 0 and 1 
```
