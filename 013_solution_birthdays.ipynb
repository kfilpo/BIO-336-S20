# Solution to the brithday problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

The logic: 

*People in a room, showing up one by one 

*Once i-th person comes into the room there are n-1 people already there 

*Calculate probability that there is no overlap in birthday, and each parson claims a date of the year

*With i-1 people, (i-1)/365 days are taken 

*The P of a new person it is 1-(i-1)/365

```python
maxn = 80             #highest number of ppl
p = np.zeros(maxn)     #future answer
p[0] = 1               #no people
p[1] = 1
for i in range(2, maxn):
    p[i] = p[i-1]*(1-(i-1)/365)
    
#the original question was about overlap so 
plt.plot(1-p);
plt.xlabel('N people in the group');
plt.ylabel('P of no shared BD');
```
